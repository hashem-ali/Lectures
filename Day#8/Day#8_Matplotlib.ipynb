```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib inline 
# this statement allows the visuals to render within your Jupyter Notebook
```

# Create a chart

```python
# Create a chart using lists, 
plt.plot([1,2,0,4,5]) # x >> index starts from 0, y >>[1,2,0,4,5]
# try to use ; sign
```

```python
x = [10,20, 30, 50, 80]
plt.plot(x, [1,2,0,4,5])
```

```python
# Multiple Lines
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
```

```python
# Create a chart using Numpy Arrays
plt.plot(np.array([9,5,3,7]))
```

```python
# Create a chart using Series
plt.plot(pd.Series([9,5,3,7]))
```

```python
# Create a chart using dataframe
plt.plot(pd.DataFrame([9,5,3,7]))
```

# Types of Charts

```python
# Line Chart
x = np.arange(0, 100) # generate a sequence of values between 0 and 99 (end value not include)
y = np.random.randint(5, size=100) # random 100 integers (from 0 to 4)
plt.plot(x, y)
```

```python
# Bar Chart
x = np.arange(10)
y = np.arange(10)
plt.bar(x, y)
```

```python
# Scatter Plot
x = np.random.randn(500)
y = np.random.randn(500)
plt.scatter(x, y)
```

```python
plt.scatter(x,y,alpha = 0.5) # alpha between 0 (transparent) and 1 (opaque)
```

```python
plt.scatter(x, y, marker = '+', s = 100) # s >> size
```

```python
# Histogram
x = np.random.randn(500) 
plt.hist(x,10); # bins = 10 means 10 bars
```

```python
# Pie Chart
plt.pie(np.arange(1,5),labels=['Jan','Feb','Mar', 'Apr']);
```

```python
plt.pie(np.arange(1,5), labels=['Jan','Feb','Mar', 'Apr'], explode = [.1,.2,.3,0]);
```

# Formatting Charts

```python
# Add title
plt.pie(np.arange(1,5),labels=['Jan','Feb','Mar', 'Apr']);
plt.title('Months', fontsize = 30, loc='center', color='b');
```

```python
# Add x and y axises
x = np.arange(10)
y = np.arange(10)
plt.bar(x, y)
plt.xlabel('Days')
plt.ylabel('Sales')
```

```python
# Add Gridlines
plt.bar(x, y)
plt.xlabel('Days')
plt.ylabel('Sales');
plt.grid()
```

```python
# Multiple Lines
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
# Add Legends
plt.legend(['2008','2009'], shadow = True, loc = 0) # location (0 = best)
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
```

```python
# Multiple Lines
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
# Add Legends
plt.legend(['2008','2009'], shadow = True, loc = 'center') 
```

```python
# Colors
# https://matplotlib.org/stable/gallery/color/named_colors.html
plt.plot([1,2,0,4,5], 'g')
plt.plot([10,9,8,6,3], c ='lime')
plt.plot([12,13, 10, 20,13], color ='blue')
```

```python
# RGB Colors
# https://matplotlib.org/stable/tutorials/colors/colors.html
plt.plot([1,2,0,4,5], '#4B0082')
plt.plot([10,9,8,6,3], c ='#029386')
plt.plot([12,13, 10, 20,13], color ='#A52A2A')
```

```python
# Line Style and width
# https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot([1,2,0,4,5], linestyle = '--', linewidth = 2, c ='#4B0082')
plt.plot([10,9,8,6,3], linestyle = ':', linewidth = 3, c ='#029386')
plt.plot([12,13, 10, 20,13],  linestyle = '-', linewidth = 1, color ='#A52A2A')
```

```python
# Figure Size
plt.figure(figsize = [10,5]) #[width, height]
plt.plot([1,2,0,4,5], linestyle = '--', linewidth = 2, c ='#4B0082')
plt.plot([10,9,8,6,3], linestyle = ':', linewidth = 3, c ='#029386')
plt.plot([12,13, 10, 20,13],  linestyle = '-', linewidth = 1, color ='#A52A2A')
```

# Subplots

```python
plt.pie(np.arange(1,5),labels=['Jan','Feb','Mar', 'Apr']);
```

```python
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.legend(['2008','2009'], shadow = True, loc = 0)
```

```python
plt.figure(figsize=[20,5])
plt.suptitle('Subplots',fontsize = 16)

#first
plt.subplot(1,2,1) #(number of rows, number of columns, number of plot)
plt.pie(np.arange(1,5),labels=['Jan','Feb','Mar', 'Apr']);
plt.title('Months')

#second
plt.subplot(1,2,2)
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.legend(['2008','2009'], shadow = True, loc = 0)
plt.title('Years');

```

```python
# Removing the Axis
plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.legend(['2008','2009'], shadow = True, loc = 0)
plt.axis('off');
```
